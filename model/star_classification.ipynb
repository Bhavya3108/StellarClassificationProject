## 6. Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x=d.drop(['class'],axis=1)
x
```

```python
y=d.loc[:,['class']]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

## 7. Feature selection

## i. Variance threshold

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vt=VarianceThreshold(threshold=0)
vt.fit(x_train)
```

```python
vt.get_support()
```

```python
x_train.columns[vt.get_support()]
```

```python
cols=[col for col in x_train.columns if col not in x_train.columns[vt.get_support()]]
cols
```

```python
x_train.drop(cols,axis=1,inplace=True)
```

```python
x_test.drop(cols,axis=1,inplace=True)
```

## ii. Correlation

```python
x_train.corr()
```

```python
plt.figure(figsize=(17,7))
sns.heatmap(x_train.corr(),annot=True,cmap='viridis',vmin=-1,vmax=1)
plt.show()
```

u,g,r,objid,specobjid,mjd can be dropped

```python
x_train.drop(['u','g','r','MJD','spec_obj_ID','obj_ID'],axis=1,inplace=True)
```

```python
x_test.drop(['u','g','r','MJD','spec_obj_ID','obj_ID'],axis=1,inplace=True)
```

## iii. Mutual information

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
mutual_info_classif(x_train,y_train)
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
sel_cols=SelectKBest(mutual_info_classif,k=8)
sel_cols.fit(x_train,y_train)
columns=x_train.columns[sel_cols.get_support()]
```

```python
columns
```

```python
cols=[col for col in x_train.columns if col not in columns]
```

```python
x_train.drop(cols,axis=1,inplace=True)
x_test.drop(cols,axis=1,inplace=True)
```

## 8. Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

## 9. Model training and testing

## i. Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
m1=LogisticRegression()
m1.fit(x_train,y_train)
y1=m1.predict(x_test)
accuracy_score(y_test,y1)*100
```

## ii. Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
m2=DecisionTreeClassifier()
m2.fit(x_train,y_train)
y2=m2.predict(x_test)
accuracy_score(y_test,y2)*100
```

## iii. Svm

```python
from sklearn.svm import SVC
```

```python
m3=SVC()
m3.fit(x_train,y_train)
y3=m3.predict(x_test)
accuracy_score(y_test,y3)*100
```

## iv. Knn

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
m4=KNeighborsClassifier()
m4.fit(x_train,y_train)
y4=m4.predict(x_test)
accuracy_score(y_test,y4)*100
```

## v. Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
m5=RandomForestClassifier()
m5.fit(x_train,y_train)
y5=m5.predict(x_test)
accuracy_score(y_test,y5)*100
```
